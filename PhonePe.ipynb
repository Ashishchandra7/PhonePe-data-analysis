```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sqlalchemy import create_engine, text
import json, os
```

```python
agg_trans_path = "data/aggregated/transaction/country/india/state"

agg_trans_list = []

for state in os.listdir(agg_trans_path):
    state_path = os.path.join(agg_trans_path, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for file in os.listdir(year_path):
            filepath = os.path.join(year_path, file)
            with open(filepath) as f:
                data = json.load(f)
            quarter = int(file.strip(".json"))
            for t in data["data"]["transactionData"]:
                agg_trans_list.append({
                    "State": state,
                    "Year": int(year),
                    "Quarter": quarter,
                    "Transaction_type": t["name"],
                    "Transaction_count": t["paymentInstruments"][0]["count"],
                    "Transaction_amount": t["paymentInstruments"][0]["amount"]
                })

df_agg_trans = pd.DataFrame(agg_trans_list)
print(f"{len(df_agg_trans)}")
df_agg_trans.head()
```

```python
agg_insurance_path = "data/aggregated/insurance/country/india/state"

agg_insurance_list=[]

for state in os.listdir(agg_insurance_path):
    state_path = os.path.join(agg_insurance_path, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for file in os.listdir(year_path):
            filepath = os.path.join(year_path, file)
            with open(filepath) as f:
                data = json.load(f)
            quarter = int(file.strip(".json"))
            user_data = data["data"]
            registered = user_data.get("aggregated", {}).get("registeredUsers", 0)
            app_opens  = user_data.get("aggregated", {}).get("appOpens", 0)
            brands     = user_data.get("usersByDevice") or []
            if brands:
                for b in brands:
                    agg_insurance_list.append({
                        "State": state, "Year": int(year), "Quarter": quarter,
                        "Registered_users": registered, "App_opens": app_opens,
                        "Brand": b["brand"], "User_count": b["count"],
                        "User_percentage": b["percentage"]
                    })
            else:
                agg_insurance_list.append({
                    "State": state, "Year": int(year), "Quarter": quarter,
                    "Registered_users": registered, "App_opens": app_opens,
                    "Brand": None, "User_count": 0, "User_percentage": 0
                })

agg_insurance = pd.DataFrame(agg_insurance_list)
print(f"{len(agg_insurance)}")
agg_insurance.head()
```

```python
map_user_path = "data/map/user/hover/country/india/state"

map_user_list = []

for state in os.listdir(map_user_path):
    state_path = os.path.join(map_user_path, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for file in os.listdir(year_path):
            filepath = os.path.join(year_path, file)
            with open(filepath) as f:
                data = json.load(f)
            quarter = int(file.strip(".json"))
            
            # hoverData is a dict, not a list — loop through key-value pairs
            hover_data = data["data"]["hoverData"]
            for district_name, values in hover_data.items():
                map_user_list.append({
                    "State": state, 
                    "Year": int(year), 
                    "Quarter": quarter,
                    "District": district_name,
                    "Registered_users": values["registeredUsers"],
                    "App_opens": values["appOpens"]
                })

df_map_user = pd.DataFrame(map_user_list)
print(f"{len(df_map_user)}")
df_map_user.head()
```

```python
top_user_path = "data/top/user/country/india/state"

top_user_dist_list = []
top_user_pin_list  = []

for state in os.listdir(top_user_path):
    state_path = os.path.join(top_user_path, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for file in os.listdir(year_path):
            filepath = os.path.join(year_path, file)
            with open(filepath) as f:
                data = json.load(f)
            quarter = int(file.strip(".json"))

            # Districts
            districts = data["data"]["districts"] or []
            for d in districts:
                top_user_dist_list.append({
                    "State": state, "Year": int(year), "Quarter": quarter,
                    "District": d["name"],
                    "Registered_users": d["registeredUsers"]
                })

            # Pincodes
            pincodes = data["data"]["pincodes"] or []
            for p in pincodes:
                top_user_pin_list.append({
                    "State": state, "Year": int(year), "Quarter": quarter,
                    "Pincode": p["name"],
                    "Registered_users": p["registeredUsers"]
                })

df_top_user_dist = pd.DataFrame(top_user_dist_list)
df_top_user_pin  = pd.DataFrame(top_user_pin_list)

print(f"Top User Districts → {len(df_top_user_dist):,} rows")
print(f"Top User Pincodes  → {len(df_top_user_pin):,} rows")
```

```python
tables = {
    "aggregated_transaction": df_agg_trans,
    "aggregated_user":        df_agg_user,
    "map_transaction":        df_map_trans,
    "map_user":               df_map_user,
    "map_insurance":          df_map_ins,
    "top_transaction":        df_top_trans,
    "top_user_district":      df_top_user_dist,
    "top_user_pincode":       df_top_user_pin,
    "top_insurance_district": df_top_ins_dist,
    "top_insurance_pincode":  df_top_ins_pin,
}

for table_name, df in tables.items():
    df.to_sql(table_name, con=engine, if_exists="replace", index=False)

print("\n All 10 tables inserted successfully!")
```

```python
query1 = """
    SELECT "State",
           SUM("Transaction_amount") AS total_amount,
           SUM("Transaction_count")  AS total_count
    FROM   aggregated_transaction
    GROUP  BY "State"
    ORDER  BY total_amount DESC
    LIMIT  10;
"""
df_q1 = pd.read_sql(query1, engine)
df_q1
```

```python
query2 = """
    SELECT "Transaction_type",
           SUM("Transaction_count")  AS total_count,
           SUM("Transaction_amount") AS total_amount
    FROM   aggregated_transaction
    GROUP  BY "Transaction_type"
    ORDER  BY total_count DESC;
"""
df_q2 = pd.read_sql(query2, engine)
df_q2
```

```python
query3 = """
    SELECT "Year",
           SUM("Transaction_amount") AS total_amount,
           SUM("Transaction_count")  AS total_count
    FROM   aggregated_transaction
    GROUP  BY "Year"
    ORDER  BY "Year";
"""
df_q3 = pd.read_sql(query3, engine)
df_q3
```

```python
query4 = """
    SELECT "District",
           "State",
           SUM("Transaction_amount") AS total_amount
    FROM   map_transaction
    GROUP  BY "District", "State"
    ORDER  BY total_amount DESC
    LIMIT  10;
"""
df_q4 = pd.read_sql(query4, engine)
df_q4
```

```python
query5 = """
    SELECT "State",
           SUM("Registered_users") AS total_users,
           SUM("App_opens")        AS total_app_opens
    FROM   map_user
    GROUP  BY "State"
    ORDER  BY total_users DESC
    LIMIT  10;
"""
df_q5 = pd.read_sql(query5, engine)
df_q5
```
